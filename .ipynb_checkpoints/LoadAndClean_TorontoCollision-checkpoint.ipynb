```python
import mysql.connector
import pandas as pd
import csv

# Load and filter CSV
currentRow = []
with open('/path/to/TorontoCollisions.csv', 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        year = row.get('OCC_YEAR')
        if year and year.isdigit() and 2014 <= int(year) <= 2021:
            currentRow.append({
                'Date': row.get('OCC_DATE'),
                'Year': int(year),
                'Month': int(row.get('OCC_MONTH', 0)),
                'DayofWeek': row.get('OCC_DOW'),
                'CollisionHour': int(row.get('OCC_HOUR', 0)),
                'NoofFatalities': int(row.get('FATALITIES', 0)),
                'IsInjuryCollsn': int(row.get('INJURY_COLLISIONS', 0)),
                'IsPropertyDamage': int(row.get('PROP_DAMAGE', 0)),
                'IsPedestrain': int(row.get('PEDESTRIAN', 0))
            })

torontoDF = pd.DataFrame(currentRow)

# Connect to MySQL
conn = mysql.connector.connect(
    user='student',
    password='Bi3KSjqgrNOOL',
    host='localhost',
    database='student'
)
cursor = conn.cursor()

# Prepare insert query
sql = """
INSERT INTO TorontoCollisions (
    Date, Year, Month, DayofWeek, CollisionHour,
    NoofFatalities, IsInjuryCollsn, IsPropertyDamage, IsPedestrain
) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)
"""

# Convert DataFrame to list of tuples
data = [tuple(row) for row in torontoDF.to_numpy()]

# Insert in transaction with retry logic
try:
    conn.start_transaction()
    for i in range(0, len(data), 1000):  # Chunked insert to reduce lock contention
        cursor.executemany(sql, data[i:i+1000])
    conn.commit()
except mysql.connector.Error as err:
    print("Transaction failed:", err)
    conn.rollback()
finally:
    cursor.close()
    conn.close()
```
